```python
import numpy as np
import cv2
from PIL import Image
from IPython.display import display
```

```python
img = cv2.imread("/content/image.jpg")
img = cv2.resize(img,(200,300))
sketch_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
display(Image.fromarray(sketch_img))
```

```python
greyscale_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
display(Image.fromarray(greyscale_img))
```

```python
neg_img = cv2.bitwise_not(sketch_img)
display(Image.fromarray(neg_img))
```

```python
smooth_img = cv2.GaussianBlur(neg_img, (5,5),0,0)
display(Image.fromarray(smooth_img))
```
